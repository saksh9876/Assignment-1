Q3. Explain the use of the following operators using an example:
(i) /
(ii) %
(iii) //
(iv) **





ANS:
    
    
    Certainly! These are arithmetic operators in Python, and each of them serves a specific purpose when working with numbers. Let's go through each one with examples:

(i) **/** (Division Operator)**: This operator is used to perform division between two numbers. It returns the quotient (result of division).

Example:
```python
result = 10 / 3
print(result)  # Output: 3.3333333333333335
```

(ii) **% (Modulus Operator)**: The modulus operator calculates the remainder of the division between two numbers.

Example:
```python
remainder = 10 % 3
print(remainder)  # Output: 1
```

(iii) **// (Floor Division Operator)**: Floor division performs division and then rounds down the result to the nearest integer. It returns the quotient as an integer.

Example:
```python
quotient = 10 // 3
print(quotient)  # Output: 3
```

(iv) **\*\* (Exponentiation Operator)**: The exponentiation operator raises the left operand to the power of the right operand.

Example:
```python
result = 2 ** 3
print(result)  # Output: 8
```

In these examples:
- The division operator `/` divides one number by another, yielding a float.
- The modulus operator `%` calculates the remainder after division.
- The floor division operator `//` performs division and rounds the result down to an integer.
- The exponentiation operator `**` raises a number to a given power.

Remember that the behavior of these operators can vary depending on the types of the operands. For instance, using these operators with integers will produce different results than when using them with floats.

Q4. Create a list of length 10 of your choice containing multiple types of data. Using for loop print the
element and its data type.



ANS:
    
    
   Sure, here's an example of how you can create a list with various data types and use a `for` loop to print each element along with its data type:

```python
my_list = [1, 2.5, "Hello", True, [1, 2, 3], (4, 5, 6), {"a": 1, "b": 2}, None, 3.14159, "World"]

for item in my_list:
    print("Element:", item, "| Data Type:", type(item))
```

In this example, the `my_list` contains elements of different types: integers, floats, strings, booleans, lists, tuples, dictionaries, `None`, and more. The `for` loop iterates over each element in the list and uses the `type()` function to determine the data type of each element. The loop then prints each element along with its corresponding data type. 




Q5. Using a while loop, verify if the number A is purely divisible by number B and if so then how many
times it can be divisible.




ANS:
    
    
    Certainly! Here's an example of how you can use a `while` loop to verify if number `A` is purely divisible by number `B`, and if so, calculate how many times it can be divided:

```python
A = int(input("Enter the dividend (A): "))
B = int(input("Enter the divisor (B): "))

if B == 0:
    print("Division by zero is not allowed.")
else:
    quotient = 0
    while A >= B:
        A -= B
        quotient += 1
    
    if quotient > 0:
        print(f"{A+B} is not divisible by {B}.")
    else:
        print(f"{A+B} is divisible by {B}.")
        print(f"{A+B} can be divided {quotient} times.")
```

In this example:
- We input the dividend (`A`) and divisor (`B`) from the user.
- We use a `while` loop to repeatedly subtract the divisor (`B`) from the dividend (`A`) until `A` becomes less than `B`.
- The loop increments the `quotient` counter each time a subtraction is performed.
- If the `quotient` is greater than 0, then the dividend (`A`) is not divisible by the divisor (`B`).
- If the `quotient` is 0, then the dividend (`A`) is divisible by the divisor (`B`), and we print how many times it can be divided.

Note that in this code, we're assuming positive integer values for both `A` and `B`. Division by zero is also handled to avoid errors.
